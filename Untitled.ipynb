```python
#Import Dependencies
import pandas as pd
import os
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import savefig
from sqlalchemy import create_engine
from config import password
```

```python
#Define URL and create engine
db_url = (f"postgresql+psycopg2://postgres:{password}@localhost:5432/SQL_HW")
engine = create_engine(db_url)
connection = engine.connect()
```

```python
#Import SQL Database into Pandas DataFrame
salary_data = pd.read_sql("SELECT * FROM salaries", connection)
max_sal = salary_data["salary"].max()
min_sal = salary_data["salary"].min()
salary_data.head(5)
```

```python
#Create a histogram to visualize the most common salary range for employees
salary_bins = list(range(30000, 140000, 10000))
nameobins = ["30-40k","40-50k","50-60k","60-70k","70-80k","80-90k","90-100k", "100-110k", "110-120k", "120-130k", ">130k"]
fig, ax = plt.subplots(figsize =(20, 10))
ax.hist(salary_data["salary"],bins = salary_bins)
plt.title("Salary Ranges at Pewlett Hackard")
plt.ylabel("Number of Employees")
plt.xlabel("Salary Ranges")
ax.set_xticks(salary_bins)
ax.set_xticklabels(nameobins)
savefig('Visualizations/salary_ranges.png')
plt.show()
```

```python
#Create a bar chart of average salary by title

#Organize data slice for visualization
avg_salary = pd.read_sql("SELECT titles.title, AVG(salaries.salary) \
FROM titles, salaries, emps \
WHERE titles.title_id = emps.emp_title \
AND salaries.emp_no = emps.emp_no \
GROUP BY titles.title", connection)
avg_salary
```

```python
#Create a bar chart of average salary by title
plt.subplots(figsize =(15, 7))
plt.bar(avg_salary["title"],avg_salary["avg"])
plt.xlabel("Job Title")
plt.ylabel("Salary ($)")
plt.title("Average Salary vs Job Title")
savefig('Visualizations/avg_salaries.png')
plt.show()
```
